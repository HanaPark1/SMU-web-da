```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../Day5/data/score.xlsx', index_col='지원번호')
df
```

```python
df.rename(columns={'이름':'성명'},inplace=True)
df
```

```python
df.rename(columns={'성명':'name','학교':'school'},inplace=True)
df
```

```python
df.columns = ['name', 'school', 'height', 'kor', 'eng', 'math', 'science', 'social', 'SW']
df
```

```python
# 학교 - 가산고등학교, 디지털고등학교 변경
df['학교'] = df['학교'] + '등학교'

# SW특기 소문자로 변경
df['SW특기'] = df['SW특기'].str.lower()

# NaN데이터 없음으로 변경
df.fillna('없음',inplace=True)
```
