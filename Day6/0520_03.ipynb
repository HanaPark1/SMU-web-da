```python
import pandas as pd
import numpy as np

df = pd.read_excel('../Day5/data/score.xlsx')
```

```python
## 인덱스 이름으로 지정하고 이름으로 역순 정렬
df.reset_index()
df.set_index('이름',inplace=True)
df.sort_index(ascending=False,inplace=True)
df
```

```python
# 컬럼 정렬 (순차, 역순)

df.sort_values('이름',inplace=True)
df.sort_values('이름',ascending=False,inplace=True)
df
```

```python
## 국어 수학 점수로 역순 
df.sort_values(['국어','수학'], ascending=[False,True], inplace=True)
df
```
