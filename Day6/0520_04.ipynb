```python
import pandas as pd
import numpy as np

df = pd.read_excel("../Day5/data/score.xlsx")
```

```python
## 데이터 수정
## 디지털고를 소프트고로 변경하시오
## 특정 데이터 찾아서 변경, replace
df['학교'].replace({'디지털고':'소프트고'},inplace=True)
df

## 2개를 변경
df['학교'].replace({'가산고':'금천고','디지털고':'단지고'}, inplace=True)
df
```

```python
## 소문자로 변경
df['SW특기'] = df['SW특기'].str.lower()
df

## 대문자로 변경
df['SW특기'] = df['SW특기'].str.upper()

## 첫글지 대문자, 나머지 소문자로 변경
df['SW특기'] = df['SW특기'].str.capitalize()

## 학교 -> 가산고등학교, 디지털고등학교
df['학교'] = df['학교'] + '등학교'
```

```python
### 컬럼 추가
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] +df['사회']

## 평균을 추가하시오
df['평균'] = df['합계'] / 5

## 결과를 추가하시오.
df['결과'] = '불합격'

##평균 컬럼이 60점 이상인 학생만 출력
filt = df['평균'] >= 60
df[filt]

## 평균 60 이상일 시 결과 컬럼에 합격이라 입력
df['결과'] = '합격'
```

```python
df = pd.read_csv('../Day5/data/stuscore.csv', index_col='번호')

df['합계'] = df['국어'] + df['영어'] + df['수학']
df['평균'] = df['합계'] / 3
df['결과'] = '불합격'
```
