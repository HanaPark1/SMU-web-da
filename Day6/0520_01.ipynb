```python
##조건식

filt = df['키'] > 185
df[filt]
```

```python
# & | ~
df[df['키']>=190]
df[(df['키'] >= 190) & (df['국어'] >= 90)]
df[(df['키'] >= 190) | (df['국어'] >= 90)]
df[~(df['키']>=190)]
```

```python
### str 함수
# '홍' 포함 결과 찾기
filt = df['이름'].str.contains('홍')
df[filt]

# 마지막 글자 '순' 찾기
filt = df['이름'].str.endswith('순')
df[filt]
```

```python
arr = ['python','java']
df['SW특기'].str.lower() #소문자 변경
df['SW특기'].str.upper() #대문자 변경
filt = df['SW특기'].isin(arr) # 리스트 타입을 가지고 일치하는 것을 검색
df[filt]
df
```
