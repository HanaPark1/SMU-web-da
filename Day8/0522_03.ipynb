```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 폰트 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False #그래프 마이너스 표시
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
# plt.bar(x,y)
plt.title("그래프")
```

```python
plt.plot([-1,0,1,2], [-5,-1,1,2])
plt.show()
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y) # 축데이터 지정
plt.title("온도 조절 그래프") # 그래프 타이틀
plt.xlabel("x - 월별", color='red') # x축 설명문구
plt.ylabel("y - 온도", color='blue') # y축 설명문구
plt.xticks([1,2,3])
plt.yticks([1,2,4,8])
plt.show() # 그래프 출력
```
