```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 폰트 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False #그래프 마이너스 표시
```

```python
x = [1,2,3,4,5]
y = [10,20,22,25,29]

## 꺾은선그래프, title 월별매출액, x - 월별 | 색 그린, y - 매출액 | 색 레드

plt.plot(x,y)
plt.title("월별매출액")
plt.xlabel("월별", color="green")
plt.ylabel("매출액", color="red")
plt.xticks([1,1.5,2,2.5,3,3.5,4,4.5,5])
plt.yticks([10,20,30])
plt.show()
```
