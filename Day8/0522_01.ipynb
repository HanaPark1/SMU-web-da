```python
import pandas as pd
import numpy as np
```

```python
data = {
    '이름':['홍길동','유관순','이순신','강감찬','김구','김유신','홍길자','홍길순'],
    '원':['19,700','18,400','16,800','18,700','18,800','20,200','18,800','19,000'],
    '우대':[90,40,80,40,15,80,55,100]
}
```

```python
df = pd.DataFrame(data)
df

## 최종 컬럼 생성 원+원 * 우대 * 0.01
df['원'] = df['원'].str.replace(',','').astype(int)

df['최종'] = df['원']+df['원']*df['우대']*0.01
df
```

```python
def change(data):
    if data['우대'] >= 50:
        return data['원'] + data['원'] * data['우대'] * 0.01
    else:
        return data['원']
    
df['최종'] = df.apply(change, axis=1) #row행 데이터 모두 전달
df
```
