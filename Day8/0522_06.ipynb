막대그래프

```python
name = ['홍길동', '유관순', '이순신']
```

```python
colors = ['r','g','b']
plt.bar(name,kor, label='국어',color=colors)
# color: 막대그래프 색상
# alpha : 투명도 지정
plt.yticks([70,80,90,100]) # 눈금표시 (범위 지정 x)
plt.ylim(70,105) # 눈금 범위 지정
plt.legend()
plt.show()
```

```python
colors = ['r','g','b']
plt.bar(name,kor, label='국어',color=colors, alpha=0.7)
# color: 막대그래프 색상
# alpha : 투명도 지정
plt.xticks(rotation=90) #x축 글자 90도 돌림
plt.yticks([70,80,90,100]) # 눈금표시 (범위 지정 x)
plt.ylim(70,105) # 눈금 범위 지정
plt.legend()
plt.show()
```
