```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../Day5/data/score.xlsx', index_col='지원번호')
```

```python
# # int타입에 cm문자 추가 방법
# df['키'] = df['키'].astype(str) + 'cm'

# 함수 사용해서 추가
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm)
df['합계'] = df['국어'] + df['영어'] +df['수학']+df['과학']+df['사회']
df['평균'] = df['합계'] / 5
df['등급'] = 'F'
df

def rank(score):
    result = ''
    if score >= 90 :
        result = 'A'
    elif score >= 80:
        result = 'B'
    elif score >= 70:
        result = 'C'
    elif score >= 60:
        result = 'D'
    else:
        result = 'F'
    return result

df['등급'] = df['평균'].apply(rank)
df
```

```python
# df['SW특기'].str.lower()
# df['SW특기'].str.capitalize()

# na : True NaN데이터 포함 검색, False:제외
filt = df['SW특기'].str.lower().str.contains('java', na=False)
df[filt]
```

```python
### SW특기 컬럼 글자를 전부 대문자로 변경, NaN데이터는 C를 입력
df['SW특기'].fillna("C", inplace=True)
df['SW특기'] = df['SW특기'].str.upper()


def changeData(a_data):
    return a_data.str.upper()

df['SW특기'] = df['SW특기'].apply(changeData)
```
