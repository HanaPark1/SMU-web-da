```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../Day5/data/score.xlsx', index_col='지원번호')
```

```python
df
```

```python
## numpy 명령어 np.nan NaN 데이터로 변경
df['학교'] = np.nan
```

```python
df
```

```python
### 모든 NaN데이터에 적용
df.fillna('', inplace=True)
df['학교'].fillna('가산고')
```

```python
## dropna : NaN 삭제 - row 기준으로 삭제
# df.dropna() # row 기준으로 삭제
df.dropna(axis='index', how='any') ## axis 기준: index, column | how: all,any
df.dropna(axis='columns', how='all')
```

```python
## 데이터 정렬 sort_index, sort_values

df.sort_index(ascending=False) # ascending=False 역순 정렬 속성

df.reset_index(inplace=True) # drop=True, inplace=True
df.set_index('이름', inplace=True)
df.sort_index(ascending=False)
```

```python
# 각 컬럼별 정렬
# sort_values
df.sort_values('키') #.head() 5개 정렬, inplace=True
df.sort_values('키', ascending=False)
df.sort_values(['국어', '수학']) # 국어로 먼저 정렬 후에도 같은 건 수학으로 정렬
```

```python
### 데이터 수정
## 1. replace - 컬럼 반영하지 않아도 자동으로 반영
df.replace ({'가산고':'금천고', '디지털고':'단지고'}, inplace=True) #inplace=True

## 2. 컬럼 직접수정
df['학교'] = df['학교'] + '등학교'
df.loc['1번', 'SW특기'] = 'Java'


#3. str 함수 사용해서 변경
df['SW특기'] = df['SW특기'].str.upper()

## 키: int + cm: str =>: 타입이 다르면 + 연산자 사용 X
df['키'] = df['키'] + 'cm'
df['키'] = df['키'].astype(str) + 'cm'
df.info

# 4. 컬럼 두개 이상 수정
```

```python
## 유관순 키 187
df.loc['2번', '키'] = 187
df
```

```python
### 컬럼 추가 : 없는 컬럼을 입력하면 컬럼이 추가됨
df['합계'] = 0
df['합계'] = df['국어'] + df['수학'] + df['영어'] + df['과학'] + df['사회']
```

```python
## 컬럼 추가: 조건에 맞는 형태 입력
## 합계 400점 이상 출력
filt = df['합계'] >= 400
# df[filt]
# df[df['합계'] > 400]

# 결과 컬럼 추가
df['결과'] = '불합격'

# 조건에 맞는 컬럼에 데이터 수정
df.loc[filt,'결과'] = '합격'
df.loc['1번','키'] = 190
df
```

```python
# 데이터 삭제
df.drop(columns=['합계', '결과']) # inplace=True
```

```python
## 400점 미만 index 출력
filt = df['합계'] < 400
df[filt].index

## 해당 index 검색 삭제
df.drop(index=df[filt].index)
```
